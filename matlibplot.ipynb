```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,5,0.1)
y = np.exp(x)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, s=50, c="red", edgecolor="None")
plt.show()
```

```python
plt.figure(figsize=(14,14))
plt.subplot(2,2,1)
plt.scatter(x[:-1:2], y[1::2], s=50, c="red", edgecolor="None")
plt.subplot(2,2,2)
plt.legend(loc="upper left")
plt.plot(x, y, color="purple")
```

```python
plt.scatter(x, y, s=50, c="red", edgecolor="None")
plt.title("This is a Scatter Plot")
plt.xlabel("Number of Bottles [$x$]")
plt.ylabel("Level of Drunkenness [$y = e^x$]")
plt.show()
```

```python
plt.fill_between(x,np.zeros(len(x)), np.cos(x))
plt.xlabel("echs")
plt.ylabel("cos(echs)")
plt.grid()
plt.minorticks_on()
plt.xticks(np.linspace(0,5,11))
plt.show()
```
